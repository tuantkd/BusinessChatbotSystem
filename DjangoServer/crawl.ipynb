```python
import json
from bs4 import BeautifulSoup
import requests

requests.packages.urllib3.disable_warnings(requests.packages.urllib3.exceptions.InsecureRequestWarning)

# Send an HTTP request and get the page content
url = 'https://dangkykinhdoanh.gov.vn/vn/Pages/NganhNghe.aspx'
response = requests.get(url, verify=False)
html_content = response.content

# Parse the HTML content with Beautiful Soup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table in the HTML (replace 'table_id' with the actual ID or class of your table)
table = soup.find('table')

industry = []
# Check if the table is found
if table:
    # Iterate through each row in the table
    for row in table.find_all('tr'):
        # Find all the columns (td) in the current row
        columns = row.find_all('td')

        level = columns[0].text.strip()
        name = columns[5].text.strip()
        industry.append(f"{level}-{name.lower()}")
        print(f"{level}-{name}")

with open('sql_data/industry.json', 'w', encoding='utf-8') as json_file:
    json.dump(industry, json_file, indent=2, ensure_ascii=False)
        
            


```
